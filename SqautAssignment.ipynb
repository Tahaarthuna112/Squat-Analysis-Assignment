```python
import pandas as pd

# Define file paths
file_paths = {
    "3d_distances": "/content/3d_distances.csv",
    "angles": "/content/angles.csv",
    "labels": "/content/labels.csv",
    "landmarks": "/content/landmarks.csv",
    "xyz_distances": "/content/xyz_distances.csv",
}

# Load all datasets
dataframes = {name: pd.read_csv(path) for name, path in file_paths.items()}

# Print dataset names and their shapes
for name, df in dataframes.items():
    print(f"{name}: {df.shape}")
    display(df.head())  # Shows first 5 rows
```

```python
# Check missing values for each dataset
for name, df in dataframes.items():
    print(f"Missing values in {name}:")
    print(df.isnull().sum())
    print("-" * 50)
```

```python
# Start with labels as the base
merged_df = dataframes["labels"]

# Merge other datasets using 'pose_id'
for name, df in dataframes.items():
    if name != "labels":
        merged_df = merged_df.merge(df, on="pose_id", how="inner")

# Drop 'pose_id' column as it's not needed for training
merged_df.drop(columns=["pose_id"], inplace=True)

# Check merged dataset shape and preview
print(f"Merged dataset shape: {merged_df.shape}")
display(merged_df.head())
```

```python
# Separate features (X) and labels (y)
X = merged_df.drop(columns=["pose"])  # Features
y = merged_df["pose"]  # Labels
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # Normalize features
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)  # Convert text labels to numbers

# Print class mappings
class_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
print("Class Mapping:", class_mapping)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)

print(f"Training Samples: {X_train.shape[0]}, Testing Samples: {X_test.shape[0]}")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Define the model
model = keras.Sequential([
    layers.Dense(128, activation="relu", input_shape=(X_train.shape[1],)),
    layers.Dropout(0.3),  # Prevent overfitting
    layers.Dense(64, activation="relu"),
    layers.Dropout(0.3),
    layers.Dense(32, activation="relu"),
    layers.Dense(len(class_mapping), activation="softmax")  # Output layer
])

# Compile the model
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

# Summary of the model
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=30, batch_size=32, validation_data=(X_test, y_test))
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_acc * 100:.2f}%")
```

```python
import cv2
import numpy as np
import mediapipe as mp
import onnxruntime as ort
from moviepy.editor import VideoFileClip

# Load ONNX model
onnx_model_path = "squat_recognition.onnx"
session = ort.InferenceSession(onnx_model_path, providers=['CPUExecutionProvider'])
input_name = session.get_inputs()[0].name

# Initialize MediaPipe Pose
mp_pose = mp.solutions.pose
pose = mp_pose.Pose()

# Function to extract keypoints
def extract_keypoints(frame):
    image_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = pose.process(image_rgb)
    if results.pose_landmarks:
        keypoints = np.array([[lm.x, lm.y, lm.z] for lm in results.pose_landmarks.landmark]).flatten()
    else:
        keypoints = np.zeros(33 * 3)  # 33 keypoints, 3 coordinates each (x, y, z)
    return keypoints

# Process video
input_video = "squat_video.mp4"
output_video = "squat_annotated.mp4"
cap = cv2.VideoCapture(input_video)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
out = cv2.VideoWriter(output_video, cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    keypoints = extract_keypoints(frame).reshape(1, -1).astype(np.float32)
    prediction = session.run(None, {input_name: keypoints})[0]
    label = np.argmax(prediction)  # Get class with highest probability

    # Overlay prediction text
    label_text = "Correct Squat" if label == 1 else "Incorrect Squat"
    color = (0, 255, 0) if label == 1 else (0, 0, 255)
    cv2.putText(frame, label_text, (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)

    out.write(frame)

cap.release()
out.release()
cv2.destroyAllWindows()

# Convert video format for Colab compatibility
clip = VideoFileClip(output_video)
clip.write_videofile("squat_final.mp4", codec="libx264")

print("✅ Squat analysis complete! Check 'squat_final.mp4'")
```

```python
import cv2
import numpy as np
import onnxruntime as ort
import mediapipe as mp
from moviepy.editor import VideoFileClip

# Load ONNX model
onnx_model_path = "squat_recognition.onnx"
session = ort.InferenceSession(onnx_model_path, providers=["CPUExecutionProvider"])
input_name = session.get_inputs()[0].name

# Initialize MediaPipe Pose
mp_pose = mp.solutions.pose
pose = mp_pose.Pose()

# Function to extract keypoints
def extract_keypoints(image):
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    results = pose.process(image_rgb)

    if results.pose_landmarks:
        keypoints = np.array([[lm.x, lm.y, lm.z] for lm in results.pose_landmarks.landmark]).flatten()
    else:
        keypoints = np.zeros(99)  # If no landmarks detected, return zero array

    # Ensure keypoints match model input size (170)
    if keypoints.shape[0] != 170:
        keypoints = np.pad(keypoints, (0, 170 - keypoints.shape[0]), mode='constant')

    return keypoints

# Function to process each frame
def process_frame(frame):
    frame = frame.copy()  # ✅ Make frame writable

    keypoints = extract_keypoints(frame).reshape(1, -1).astype(np.float32)
    prediction = session.run(None, {input_name: keypoints})[0]
    label = np.argmax(prediction)

    # Assign labels
    labels = ["Good Squat", "Knees Too Forward", "Shallow Squat", "Back Rounding"]
    label_text = labels[label] if label < len(labels) else "Unknown"

    # Color coding
    color = (0, 255, 0) if label == 0 else (0, 0, 255)  # Green for good, Red for bad

    # ✅ Add text overlay
    cv2.putText(frame, f"Squat Analysis: {label_text}", (50, 50),
                cv2.FONT_HERSHEY_SIMPLEX, 1, color, 3)

    return frame

# Process video
input_video = "squat_video.mp4"  # Make sure this is the correct downloaded video file
output_video = "squat_analysis.mp4"

clip = VideoFileClip(input_video)
processed_clip = clip.fl_image(process_frame)
processed_clip.write_videofile(output_video, codec="libx264", fps=clip.fps)

print("✅ Squat analysis video saved as 'squat_analysis.mp4'")
```
